```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class DenseLayer:
    def __init__(self,num_of_inputs,num_of_neurons):
        self.W=np.random.randn(num_of_neurons,num_of_inputs).T
        self.B=np.random.randn(num_of_neurons).reshape((1,num_of_neurons))
    def forward(self,inputs):
        self.output=np.dot(inputs,self.W)+self.B
        return self.output
```

```python
class Relu:
    def forward(self,inputs):
        self.output=np.maximum(0,inputs)
        return self.output

class SoftMax:
    def forward(self,inputs):
        print(inputs)
        normalized_inputs=inputs-np.max(inputs,axis=1,keepdims=True)
        print(normalized_inputs)
        exp_input=np.exp(normalized_inputs)
        print(exp_input)
        self.output=exp_input/(np.sum(exp_input,axis=1,keepdims=True))
        return self.output
```

```python

def create_spiral_data(samples, classes):
    X = np.zeros((samples*classes, 2))
    y = np.zeros(samples*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(samples*class_number, samples*(class_number+1))
        r = np.linspace(0.0, 1, samples)
        t = np.linspace(class_number*4, (class_number+1)*4, samples) + np.random.randn(samples)*0.2
        X[ix] = np.c_[r*np.sin(t*2.5), r*np.cos(t*2.5)]
        y[ix] = class_number
    return X, y
```

```python
X,y=create_spiral_data(100,3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='brg')
plt.show()
```

```python
## complete forward pass
input_layer=DenseLayer(2,3)
relu_activation=Relu()
output_layer=DenseLayer(3,3)
softmax_activation=SoftMax()
input_layer.forward(X)
relu_activation.forward(input_layer.output)
output_layer.forward(relu_activation.output)
softmax_activation.forward(output_layer.output)
```

```python
class CrossEntrophyLoss:
    def forward(self,inputs,labels):
        epsilon = 1e-7
        inputs=np.log(np.clip(inputs,epsilon,1-epsilon))
        if labels.ndim>1:
            hadamard_prod=inputs*labels
            self.output=-np.sum(hadamard_prod,axis=1,keepdims=True)
        else:
            y_pred=inputs[:,y]
            self.output=-y_pred.reshape(-1,1)
        return self.output

class MeanSquaredError:
    def forward(self,y_pred,y):
        return np.mean(np.pow(y-y_pred,2),axis=0)
            
```

```python
loss=CrossEntrophyLoss()
loss.forward(softmax_activation.output,y)
```
